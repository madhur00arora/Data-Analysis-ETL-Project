```python
import pandas as pd
```

```python
#raed data from file and handle null values
df=pd.read_csv(r'C:\Users\komal\.cache\kagglehub\datasets\ankitbansal06\retail-orders\versions\4\orders.csv',na_values=['Not Available','unknown'])
#df['Ship Mode'].unique()
```

```python
df.head(5)
```

```python
#rename column names -format to lower case and _ between any two words of column name
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
#derive new columns discount, sale_price and profit
df['discount']=df['discount_percent']*df['list_price']/100
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
```

```python
df.head()
```

```python
df.dtypes
```

```python
#convert data type of order date from object to datetime
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df.dtypes
```

```python
#drop unncessary columns
df=df.drop(columns=['cost_price','discount_percent','list_price'])
```
